### Text/Document Preprocessing

#### Data cleaning and preprocessing

#### Cleaning the texts

**NLTK**: Natural Language Processing Toolkit is a python library that is used for performing all the NLP tasks like stemming, lemmatizing or removing stopwords, etc.

**Porter Stemme**r: It is a type of stemmer that is used for stemming. stemming is basically a technique of converting a word to its root word.
Ex: learning → learn

```python
import re  ## used to perform regex in python.
import nltk  # Preprocessing our text
# nltk.download('stopwords')
from nltk.corpus import stopwords  # removing all the stop words
from nltk.stem.porter import PorterStemmer   # stemming of words
from nltk.stem import WordNetLemmatizer  # Lemmitization of words
# nltk.download('wordnet')


corpus = []  ##  Empty corpus list, can be used to store all the text after cleaning.
for i in range(0, 5572):
    review = re.sub('[^a-zA-Z]', ' ', dataset['v2'][i])
    review = review.lower()
    review = review.split()
    
    all_stopwords = stopwords.words('english')
#     all_stopwords.remove('not')
    
#     Stemming
#     ps = PorterStemmer()
#     review = [ps.stem(word) for word in review if not word in set(all_stopwords)]

#     sentences = nltk.sent_tokenize(paragraph)

#     Lemmatization
    lemmatizer = WordNetLemmatizer()
#     for i in range(len(sentences)):
#         words = nltk.word_tokenize(sentences[i])
#         review = [lemmatizer.lemmatize(word) for word in review if word not in set(stopwords.words('english'))]
    review = [lemmatizer.lemmatize(word) for word in review if word not in set(all_stopwords)]

#     sentences[i] = ' '.join(words)     
    review = ' '.join(review)
    corpus.append(review)
```

```python
# print(corpus)
corpus
```

### Creating the Bag of Words model

To convert the list of the sentence(corpus) into vectors so that we can feed this data into our machine learning model. for converting the text into vectors we are going to use a bag of words which is going to convert the text into binary form.

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=3500)
X = cv.fit_transform(corpus).toarray()


# y=pd.get_dummies(dataset['v1'])
# y = dataset.iloc[:, 1].values

# y = dataset['v1'].values

dataset.v1 = dataset.v1.map({'spam':1,'ham':0})
y = dataset.v1.values
```

```python
# len(X)
```

### Dumping the CV for future use
For doing predictions of new emails we need to save both our model and count vectorizer.

```python
import pickle
filename='cv.pkl'
pickle.dump(cv , open(filename, 'wb'))
```

### Train Test Split
### Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
# X.shape[0] != y.shape[0]
```

### Training model using different classifiers (Naive bayes classifier)
#### Training the Naive Bayes model on the Training set

```python
from sklearn.naive_bayes import GaussianNB
classifier_nb = GaussianNB()
classifier_nb.fit(X_train, y_train)
score_nb = classifier_nb.score(X_test, y_test)

from sklearn.naive_bayes import MultinomialNB
classifier_mnb = MultinomialNB().fit(X_train, y_train)
score_mnb = classifier_mnb.score(X_test, y_test)

from sklearn.linear_model import LogisticRegression
classifier_lr = LogisticRegression()
classifier_lr.fit(X_train,y_train)
score_lr = classifier_lr.score(X_test, y_test)
```

## Evaluating Model

### Predicting the Test set results

```python
y_pred = classifier_mnb.predict(X_test)
# print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

### Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

### Making report of the classification

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

### Finding the best model

```python
accuracy_score = pd.DataFrame({
    "Naive Bayes" : score_nb,
    "Multinomial Naive Bayes" : score_mnb,
    "Logistic Regression" : score_lr
}, index = ['Model Used'])

accuracy_score
```

## Therefore, Multinomial Naive Bayes is best model.

### Predicting if a single msg is spam or not

#### Spam
Use our model to predict if the following msg:
"Hurray! Get a chance to win a car."
is a spam or not.

**Solution:** We just repeat the same text preprocessing process we did before, but this time with a single msg.

```python
new_review = 'Congrats! 1 year special cinema pass for 2 is yours. call 09061209465 now! C Suprman V, Matrix3, StarWars3, etc all 4 FREE! bx420-ip4-5we. 150pm. Dont miss out!'
new_review = re.sub('[^a-zA-Z]', ' ', new_review)
new_review = new_review.lower()
new_review = new_review.split()
# ps = PorterStemmer()
all_stopwords = stopwords.words('english')
# all_stopwords.remove('not')
# new_review = [ps.stem(word) for word in new_review if not word in set(all_stopwords)]
review = [lemmatizer.lemmatize(word) for word in review if word not in set(all_stopwords)]
new_review = ' '.join(new_review)
new_corpus = [new_review]
new_X_test = cv.transform(new_corpus).toarray()
new_y_pred = classifier_mnb.predict(new_X_test)
print(new_y_pred)
```

*The* message was correctly predicted as spam by our model.

## Prediction for a new email

```python
def new_review(new_review):
    new_review = new_review
    new_review = re.sub('[^a-zA-Z]', ' ', new_review)
    new_review = new_review.lower()
    new_review = new_review.split()
    ps = PorterStemmer()
    all_stopwords = stopwords.words('english')
    all_stopwords.remove('not')
    new_review = [ps.stem(word) for word in new_review if not word in   set(all_stopwords)]
    new_review = ' '.join(new_review)
    new_corpus = [new_review]
    new_X_test = cv.transform(new_corpus).toarray()
    new_y_pred = loaded_model.predict(new_X_test)
    return new_y_pred
new_review = new_review(str(input("Enter new review...")))
if new_review[0]==1:
  print("SPAM")
else :
  print("NOT SPAM")
```
